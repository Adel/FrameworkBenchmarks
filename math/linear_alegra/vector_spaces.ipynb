# Complex numbers
A complex number is an ordered pair (a,b) where:
* $a, b \in \mathbb{R}$
  * written as $z = a + b*i$
  * $a$ is the real part
  * $b$ is the imaginary part
* $i$ is the imaginary unit as $i^2 = -1$

Complex numbers extend the real number system to include solutions to equations like $x^2 + 1 = 0$

On a complex 2D plane, it is represented with the real part on the $x$ axis and the imaginary part on the $y$ axis.

## $ \mathbb{C}$
* $ \mathbb{C}$ is the set of all complex numbers.
* $ \mathbb{C} = { a + b*i: a,b \in \mathbb{R}}$

### Addition
$ (a + b*i) + (c + di) = (a+c) + (b+d)i$

### Multiplication
$ (a + bi) * (c + di) = (ac - bd) + (ad+bc)i$

## Properties of complex arithmetic

### Commutativity
$ \forall \alpha, \beta \in \mathbb{C} \\
 \alpha + \beta = \beta + \alpha$

### Associativity
$ \forall \alpha, \beta, \gamma \in \mathbb{C}, \\
 (\alpha + \beta) + \gamma = \alpha + (\beta + \gamma)$

### Identities
$\forall \lambda \in\mathbb{C}, \\
\lambda + 0 = \lambda \text{ and } \lambda * 1 = \lambda$

### Additive inverse       
$\forall \alpha \in \mathbb{C},  \\
\exists \beta \in \mathbb{C} \text{ such that } \alpha + \beta = 0$  

### Multiplicative inverse
$\forall \alpha \in \mathbb{C}, \text{ with } \alpha != 0,  \\
\exists \text{ unique } \beta \in \mathbb{C} \text { such that } \alpha * \beta = 1$

### Distributive property  
$\forall \alpha,\beta,\lambda \in \mathbb{C}, \\
\lambda ( \alpha + \beta) = \lambda\alpha + \lambda\beta$        

## Substraction and divison
$\forall \alpha, \beta \in \mathbb{C}$

* $- \alpha$ is the unique complex number such that: $ \alpha + (-\alpha) = 0$
  * $ \beta -\alpha = \beta + (-\alpha)

* for $\alpha != 0, \alpha * \frac{1}{\alpha} = 1$
  * $ \frac{\beta}{\alpha} = \beta * \frac{1}{\alpha}$

## $\mathbb{F}$: Field

* $\mathbb{F}$ stands for either $\mathbb{R}$ or $\mathbb{C}$

The symbol $\mathbb{F}$ is used to represent a field, which is a set of numbers where addition, substraction, multiplication and division (except by zero) are well defined.

## Scalars
Elements of $\mathbb{F}$ are called scalars. A scalar is a single value that can multiply vectors and matrices.

* $\alpha^m = \alpha*\alpha*...\alpha, m \text{ times}$
* $(\alpha^m)^n = \alpha^{mn}$


## $\mathbb{R}^n$ and $\mathbb{C}^n$

* $\mathbb{R^2}= \{(x,y): x,y \in \mathbb{R}\}$
* $\mathbb{R^3}= \{(x,y,z): x,y,z \in \mathbb{R}\}$

### Lists
A list of length $n$ is an ordered collection of $n$ elements. Each element can be a number or list, or object. It is also known as an `n-tuple`.

* The list of length `0` is : `()`


## $\mathbb{F}^n$

$\forall n \in \mathbb{N}, \\
\mathbb{F}^n = \{(x_1, x_2, \dots, x_n): x_k \in \mathbb{F} \text{ for } k = 1, 2, \dots, n\}$

* if $n>=4$, we cannot visualize $\mathbb{R}^n$ as a physical object.

### Addition in $\mathbb{F}^n$

Addition is defined by adding corresponding coordinates:

$(x_1,\dots,x_n) + (y_1,\dots, y_n) = (x_1 + y_1, \dots, x_n+y_n)$

#### Commutativity
$ \text{if} x,y \in \mathbb{F}^n \text{ then } x + y = y + x$

## Notation
To avoid explicit coordinates whenever possible, we can write:

$x = (x_1, \dots, x_n)$

For example, for `0`:

$0 = (0, \dots, 0)$

The context should determine which `0` (the list or the scalar) is intended.
For example, for the additive identity:

$ x + 0 = x \forall x \in \mathbb{F}^n$
